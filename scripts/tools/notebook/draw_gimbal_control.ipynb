```python
df = pd.read_csv("/home/ls/m300_trajectory/gimbal_control_info.csv")
df
```

```python
ctrl_times = df["ctrl_times"].to_numpy()
current_pos_x = df["current_pos_x"].to_numpy()
current_pos_y = df["current_pos_y"].to_numpy()
x_error = df["x_error"].to_numpy()
y_error = df["y_error"].to_numpy()
pitch_control = df["pitch_control"].to_numpy()
yaw_control = df["yaw_control"].to_numpy()
```

```python

font1 = {'family' : 'Times New Roman',
'weight' : 'bold',
'size'   : 15,
}
font2 = {'family' : 'Times New Roman',
'weight' : 'bold',
'size'   : 12,
}


fig = plt.figure(figsize=[10,6])
ax = fig.gca()
x_major_locator=MultipleLocator(1)
y_major_locator=MultipleLocator(100)
ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(y_major_locator)
plt.xlim(-0.1,9.5)

plt.plot(ctrl_times, current_pos_x, c="blue", label='Heat Position X', marker='o')
plt.plot(ctrl_times, current_pos_y, c="red", label='Heat Position Y', marker='.')
plt.plot(ctrl_times, x_error, c="green", label='X Error', marker='^')
plt.plot(ctrl_times, y_error, c="orange", label='Y Error', marker='s')

plt.plot(ctrl_times, [20]*10, c="grey", label='Boundary', linestyle='--')
plt.plot(ctrl_times, [-20]*10, c="grey",  linestyle='--')
# plt.plot(ctrl_times, pitch_control, c="black", label='pitch_control', marker='^')
# plt.plot(ctrl_times, yaw_control, c="purple", label='yaw_control', marker='s')
plt.xlabel('Control Times', font1)  
plt.ylabel('Pixels', font1)  
plt.grid()
plt.legend(prop=font2)
plt.show()
fig.savefig("/home/ls/forest_fire_thesis/single_fire_detection_system_paper/figures/experiment/plot_gimbal_control.pdf")
```
